# Import the necessary libraries

# Data Cleaning

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100     #df.isnull().sum()/df.shape[0]*100
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','').astype(float)
```

```python
df['rate'] = pd.to_numeric(df['rate'].astype(str).str.extract(r'(\d+\.\d+)')[0], errors='coerce')
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df['votes'] = df['votes'].astype(int)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

# Fill missing values

```python
df['location'] = df['location'].fillna(df['location'].mode()[0],inplace=True)
```

```python
df['rest_type'] = df['rest_type'].fillna(df['rest_type'].mode()[0],inplace=True)
```

```python
df['cuisines'].fillna('various',inplace=True)
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median(),inplace=True)
```

```python
df['dish_liked'].fillna('unknown',inplace=True)
```

```python
df.isnull().sum()
```

# EDA (Exploratory Data Analysis)

```python
df.head(2)
```

```python
df.columns = ['url','address','name','online_order','book_table','rate','votes','phone','location','rest_type','dish_liked','cuisines','approx_cost','review_list','menu_item','listed_type','listed_city']
```

# Top Cities with Most Restaurants

```python
top_cities = df['listed_city'].value_counts().head(10).reset_index()
top_cities
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='listed_city',y='count',data=top_cities,palette='Set1')
plt.title('Top 10 cities by number of Restaurants',fontsize=14)
plt.xlabel('Listed_city')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Remove rows with missing approx_cost values
df = df[df['approx_cost'].notnull()]
```

```python
df['cost_category'] = pd.cut(df['approx_cost'], bins=[0, 300, 800, 1500, np.inf], labels=['Low', 'Medium', 'High', 'Luxury'])
```

```python
print(df['cost_category'].value_counts())
```

```python
avg_rating = df.groupby('cost_category')['rate'].mean().reset_index()

sns.barplot(x='cost_category', y='rate', data=avg_rating, palette='Set1')
plt.title('Average Rating by Cost Category')
plt.ylabel('Average Rating')
plt.xlabel('Cost Category')
plt.show()
```

```python
sns.countplot(x='cost_category', data=df, palette='Set2')
plt.title('Restaurants by Cost Category')
plt.xlabel('Cost Category')
plt.ylabel('Count')
plt.show()
```

# Countplot by City and Cost Category

```python
# Compare average rating by cost category
df.groupby('cost_category')['rate'].mean().sort_values()
```

```python
# Cost category distribution by city
sns.countplot(x='listed_city', hue='cost_category', data=df)
plt.title('Distribution of Restaurants by City and Cost Category', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Cost Category')
plt.tight_layout()
plt.show()
```

# Average Rating by Cost Category

```python
avg_rating_by_cost = df.groupby('cost_category')['rate'].mean().sort_values(ascending=False)
```

```python
avg_rating_by_cost.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Cost Category')
plt.ylabel('Average Rating')
plt.xlabel('Cost Category')
plt.xticks(rotation=0)
plt.show()
```
